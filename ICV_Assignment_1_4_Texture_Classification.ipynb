The Local Binary Pattern (LBP) operator describes the surroundings of a pixel by generating a bit-code from the binary derivatives of a pixel.

a) Write a function that divides a greyscale image into equally sized non-overlapping windows and returns the feature descriptor for each window as distribution of LBP codes. For each pixel in the window, compare the pixel to each of its 8 neighbours. Convert the resulting bit-codes (base 2) to decimals (base 10 numbers) and compute their histogram over the window. Normalize the histogram (which is now a feature descriptor representing the window). Show in the report the resulting images.

```python
def ICV_greyscale_histogram(image):
    bins = np.arange(0,256) #creating empty array of pixel values from 0-255
    bins = bins.reshape(256,1).ravel()
    #creating emppty array to store pixel count values
    grey_hist = np.zeros((bins.shape[0]))

    for i in range(bins.shape[0]):
        grey_hist[i] = np.sum(image[:,:]==i)
    
    return grey_hist.ravel(),bins
```

```python
def ICV_greyscale(image):
    #create copy of original_image
    grey = np.zeros((image.shape[0],image.shape[1]))
    #using BT.601 to convert image to greysclae
    #Gray = (Red * 0.299 + Green * 0.587 + Blue * 0.114)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            grey[i][j] = image[i][j][0] * 0.299 + image[i][j][1] * 0.587 + image[i][j][2] * 0.114
    
    return grey
```

```python
#Utility function for comparing values of neighbours 
def ICV_feature_compare(x,y):
    if x>y:
        return 1
    else:
        return 0
```

```python
def ICV_feature_descriptor(image, block_size = 3, show_hist = True):
    #blocks of image
    windows = []
    for i in range(0,image.shape[0],block_size):
        for j in range(0,image.shape[1],block_size): 
            windows.append(image[i:i+block_size, j:j+block_size])
    #print(windows)
    
    feature_windows = []
    for count, window in enumerate(windows):
        padded_window = np.zeros((window.shape[0]+2,window.shape[1]+2))
        padded_window[1:-1,1:-1] = window[:,:]
        
        feature_window = window.copy()
        for i in range(0,feature_window.shape[0]):
            for j in range(0,feature_window.shape[1]):
                #calulate neighour
                n1,n2,n3,n4,n5,n6,n7,n8 = padded_window[i-1,j-1],padded_window[i-1,j],padded_window[i-1,j+1],padded_window[i,j+1],padded_window[i+1,j+1],padded_window[i+1,j],padded_window[i+1,j-1],padded_window[i,j-1]
                n1 = ICV_feature_compare(n1,feature_window[i][j])
                n2 = ICV_feature_compare(n2,feature_window[i][j])
                n3 = ICV_feature_compare(n3,feature_window[i][j])
                n4 = ICV_feature_compare(n4,feature_window[i][j])
                n5 = ICV_feature_compare(n5,feature_window[i][j])
                n6 = ICV_feature_compare(n6,feature_window[i][j])
                n7 = ICV_feature_compare(n7,feature_window[i][j])
                n8 = ICV_feature_compare(n8,feature_window[i][j])
                
                
                feature_window[i][j] = 2**n1 + 2**n2 + 2**n3 + 2**n4 + 2**n5 + 2**n6 + 2**n7 + 2**n8
        
        feature_windows.append(feature_window)
        
        if count%1000 == 0 and show_hist==True:
            plt.axis('off')
            plt.imshow(window,cmap='gray')
            plt.title(f'Original Window no:{count}')
            plt.show()
            
            plt.axis('off')
            plt.imshow(feature_window,cmap='gray')
            plt.title(f'LBP of Window no:{count}')
            plt.show()
            
            hist, bins = ICV_greyscale_histogram(feature_window)
            plt.plot(bins, hist)
            plt.title(f'Histogram of LBP no:{count}')
            plt.show()
            
            norm_hist = (hist - hist.mean()) / hist.std()
            plt.plot(bins, norm_hist)
            plt.title(f'Normalized Histogram of LBP no:{count}')
            plt.show()
            
    return feature_windows
```

Comment the results in the report. Is the global descriptor able to represent whole images of different types (e.g. faces vs. cars)?
```python
def ICV_lsbh_euclidian_distance(hist1, hist2):
    
    min_index = min(len(hist1),len(hist2))
    
    
    distance = 0
    for i in range(min_index):
        distance += (hist1[i]-hist2[i])**2
    
    return math.sqrt(distance)
```

```python
def local_binary_pattern_histogram(image, grid_x = 8, grid_y=8):
    
    concated_hist = []
    
    grid_x = image.shape[0] // grid_x
    grid_y = image.shape[1] // grid_y    
    for i in range(0,image.shape[0],grid_x):
        for j in range(0,image.shape[1],grid_y): 
            img_window = image[i:i+grid_x, j:j+grid_y]
            window_hist, bins = ICV_greyscale_histogram(img_window)
            #plt.plot(bins,window_hist,color='black')
            #plt.show()
            concated_hist.append(window_hist)
    
    concated_hist = np.array(concated_hist).ravel()
    
    #plt.plot(range(0,len(concated_hist)),concated_hist, color='black')
    #plt.show()
    
    return concated_hist

```

```python
train_face_A = cv2.imread('./DatasetA/face-1.jpg')
train_face_A = ICV_greyscale(train_face_A)

block_size = 3
train_face_lsb_A = ICV_feature_descriptor(train_face_A, block_size,show_hist = False)

train_image_A = np.zeros((train_face_A.shape[0],train_face_A.shape[1]))
count = 0
for i in range(0,train_face_A.shape[0],block_size):
    for j in range(0,train_face_A.shape[1],block_size): 
        train_face_A[i:i+block_size, j:j+block_size] = train_face_lsb_A[count]
        count+=1
train_face_lsbh_A = local_binary_pattern_histogram(train_face_A, 8, 8)


train_face_B = cv2.imread('./DatasetA/face-2.jpg')
train_face_B = ICV_greyscale(train_face_B)

block_size = 3
train_face_lsb_B = ICV_feature_descriptor(train_face_B, block_size,show_hist = False)

train_image_B = np.zeros((train_face_B.shape[0],train_face_B.shape[1]))
count = 0
for i in range(0,train_face_B.shape[0],block_size):
    for j in range(0,train_face_B.shape[1],block_size): 
        train_face_B[i:i+block_size, j:j+block_size] = train_face_lsb_B[count]
        count+=1
train_face_lsbh_B = local_binary_pattern_histogram(train_face_B, 8, 8)

threshold = ICV_lsbh_euclidian_distance(train_face_lsbh_A, train_face_lsbh_B)
print(threshold)
```
