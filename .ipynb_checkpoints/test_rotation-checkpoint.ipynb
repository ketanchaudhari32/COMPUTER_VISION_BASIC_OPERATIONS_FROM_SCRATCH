```python
#Function for image rotation
def ICV_rotation_skewing(img,angle):
    #getting image height and width
    original_img_height = img.shape[0]
    original_img_width = img.shape[1]
    
    #converting angle from degree to radians to make it possible to relate to linear measure and angle measures
    angle = math.radians(angle)
    
    #getting trignometric values of angle i.e cosine and sine
    cosine = math.cos(angle)
    sine = math.sin(angle)
    
    #finding centre of original image to perform rotation w.r.t to centre
    original_centre_height = original_img_height//2
    original_centre_width = original_img_width//2
    
    #creating empty image to store rotated image
    
    '''
    Since img will be rotated at certain angle, the dimensions of image 
    will be change as per angle of rotation.
    By using trignometric functions we can calculate new height and width
    '''
    rotated_img_height = int(original_img_width*cosine + original_img_height*sine)
    rotated_img_width = int(original_img_width*sine + original_img_height*cosine)
    rotated_img = np.zeros([rotated_img_height,rotated_img_width,3],dtype=np.uint8)
    #rotated_img = np.zeros([original_img_height,original_img_width,3],dtype=np.uint8)
    
    
    #finding centre of empty rotated image
    rotated_centre_height = (rotated_img_height)//2
    rotated_centre_width = (rotated_img_width)//2
    
    #empty_centre_height = (empty_img_height)//2
    #empty_centre_width = (empty_img_width)//2
    
    #Iteration over every pixel in original image
    for i in range(original_img_height):
        for j in range(original_img_width):
            #finding co-ordinated of rotated pixel values
            #new_x = int(abs(cosine*(i)-original_centre_height) + abs(sine*(j)-original_centre_width))
            #new_y = int(- abs(sine*(j)-original_centre_height) + abs(cosine*(i)-original_centre_width))
            
            #new_x = int(cosine*(i) - sine*(j))
            #new_y = int(sine*(j) + cosine*(i))
            
            function_matrix = np.array([[cosine, -sine],   #clockwise direction
                                        [sine, cosine]])
            
            multiple_matrix = np.array([i,j])
            
            centre_coordinate_matrix = np.array([-original_centre_height,-original_centre_width])
            
            new_cordinates = function_matrix @ multiple_matrix - centre_coordinate_matrix
            
            new_x = int(new_cordinates[0])
            new_y = int(new_cordinates[1])
            
            #print(new_x,new_y)
            
            #Adjusting co-ordinates with respect to centre of empty rotated image
            new_x = int(abs( rotated_centre_height - new_x))
            new_y = int(abs( rotated_centre_width - new_y))
            
            #updating values of rotated images
            #rotated_img[new_x][new_y][0] = img[i][j][0]
            #rotated_img[new_x][new_y][1] = img[i][j][1]
            #rotated_img[new_x][new_y][2] = img[i][j][2]
            #rotated_img[new_x, new_y, :] = img[i,j,:]
            try:
                rotated_img[new_x, new_y, :] = img[i,j,:]
            except:
                pass
            
    return np.array(rotated_img)
            
    
    
```
